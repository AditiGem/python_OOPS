```python
import re
```

```python
#1.   dit(.)
text="cat cot cut c9t c@t"
result=re.findall("c.t",text)
print(result)
```

```python
#2. start of string
text="Hello world"
print(bool (re.search("^Hello",text)))
print(bool (re.search("^world",text)))
```

```python
#3. end of string
text="Hello world"
print(bool (re.search("Hello$",text)))
print(bool (re.search("world$",text)))
```

```python
#4.  star(*) zero or more
text = "helloooo"
result = re.findall("lo*",text)
result
```

```python
#5. (+) one and more
text = "hellooo"
result=re.findall("lo+",text)
result
```

```python
#6. 0 or 1 time (?)
text = "color colour"
result = re.findall("colou?r",text)
result
```

```python
#7. character set ([])
text = "apple ball cat"
result=re.findall("[abc]",text)
result
```

```python
# regex to validate 10 digit mobile no.
import re
pattern = r'^\d{10}$'
mobile = "9876543210"
if re.match(pattern, mobile):
    print("Valid mobile number",mobile)
else:
    print("Invalid mobile number")
```

```python
str=" contact me at test@gmail.com or admin@yahoo.in"
pattern = r'[\w.-]+@[\w.-]+\.\w+'
email=re.findall(pattern,str)
print(email)
```

```python
# validate ipv4
ips = [
    "192.168.1.1",
    "255.255.255.255",
    "256.1.1.1",
    "999.10.10.10",
    "127.0.0.1",
    "192.168.1"]
pattern = r'^(\d{1,3}\.){3}\d{1,3}$'
for ip in ips:
    if re.match(pattern, ip):
        print(ip)
```
