```python
#  move all zeroes at the end 
nums=[0,1,0,3,12]
result = []# to store no zero elements
zero_count=0
for n in nums:
    if n==0:
        zero_count+=1
    else:
        result.append(n)
for i in range(zero_count):
    result.append(0)
print(result)
```

```python
text="naman"
left = 0 
right =len(text)-1
isPalindrome=True
while left<right:
    if text[left]!=text[right]:
        isPalindrome=False
        break
    left ==1
    right-=1
print("Palindrome")
```

```python
nums=[1,2,3,2]
target=5
count =0
for i in range(len(nums)):
    total = 0
    for j in range (i,len(nums)):
        total+=nums[j]
        if total == target:
            count+=1
print(count)
```

```python
#brute force approch to get sum of 2 digits which matches the target and then printing them
nums=[2,7,11,15]
target = 9
for i in range(len(nums)):
    for j in range(i+1,len(nums)):
        #check if sum matches the target
        if nums[i]+nums[j]==target:
            print(nums[i],nums[j])
```

```python
# 2 pointer approch to print sum of 2 digits innthe list to be equal to the target 
nums=[2,7,11,15]
target = 9
left =0
right =len(nums)-1
while left<right:
    correct_sum = nums[left]+nums[right]
    if(correct_sum == target):
        print(nums[left],nums[right])
        break
    elif correct_sum<target:
        left+=1
    else:
        right-=1
        
```

# Tortoise and here technique

```python
nums = [10,20,30,40,50]   # contains a cycle conceptually
slow = 0
fast = 0

while fast<len(nums) and fast+1<len(nums):
    slow += 1          # 1 step
    fast += 2   # 2 steps

print(nums[slow])
     
```
