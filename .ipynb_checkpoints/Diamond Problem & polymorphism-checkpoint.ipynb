```python
class A:
    def message(self):
        print("this is message from class")
class B:
    def message(self):
        print("this is msg from class B")
class C(A):
    def message(self):
        print("this is msg from class C")
class D(B,C):
    def message(self):
        print("this is msg from class D")
# obj=A()
# o1=B()
# o2=C()
o3=D()
# obj.message()
# o1.message()
# o2.message()
o3.message()
```

### POlYMORPHISM


```python
class Human:
    def speak(self):
        print("Human is speaking")
class Dog:
    def speak(self):
        print("dog is barking")
class Baby:
    def speak(self):
        print("baby is crying")
#polymorphism in action
h=Human()
b=Baby()
d=Dog()
h.speak()
d.speak()
b.speak()
        
```

# Abstract Class
## method name is present but the method body is missing ,
## the method is defined in the child class

```python
class Parent:
    def work(self):
        pass
class child(Parent):
    def work(self):
        print("I am working")
```

```python
class Payment:
    def pay(self,amount):
        pass
class UPI(Payment):
    def pay(self,amount):
        print("Paid using UPI:",amount)

obj =UPI()
obj.pay(12)
```
