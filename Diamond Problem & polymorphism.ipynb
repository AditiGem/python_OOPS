# Composition 
### dont rely on inheritence. use object of another class

```python
class Address:
    def __init__(self,city):
        self.city=city
    def show_add(self):
        print("city:",self.city)

class Student:
    def __init__(self,name,city):
        self.name=name
# compostion:
#creating object of adress class inside student class
        self.address=Address(city)
    def show_student(self):
        print("name:",self.name)
#using obj of another class
        self.address.show_add()
s=Student("Karan","Delhi")
s.show_student()
```

```python
class Engine:
    def start(self):
        print("Engine started")


class Car:
    def __init__(self):
        self.engine = Engine()   # HAS-A (composition)

    def start_car(self):
        print("Car is starting...")
        self.engine.start()


# Creating object of Car
my_car = Car()

# Starting the car
my_car.start_car()
```
