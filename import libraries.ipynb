```python
# math 
# this program demonstrates basic functions of the math module
import math
```

```python
#1.square root of a number
num=16
sqrt_result=math.sqrt(num)
print("Square root of num is:",sqrt_result)
```

```python
# 2.power of a number 
power_result=math.pow(2,3)
print("2 raised to power 3 is ",power_result)
```

```python
#3. Floor funcction
#return the largest integer less than or equal to the number
print("Ceil of 7.8 is",math.ceil(7.8))
```

```python
#5. Absolute value 
# returns positive value of no.
print("Absolute value of -5 is", math.fabs(-5))
```
