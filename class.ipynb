## inheritance one class uses another class properties

```python
class Human:
    def eat(self):
        print("human can eat ")
class Student(Human):
    def study(self):
        print("student can study")
s1=Student()
s1.eat()

class Animal:
    def sound(self):
        print("all animal can speak")
class Dog(Animal):
    def pet(self):
        print("all dogs can bark")
s1=Dog()
s1.sound()
```

```python
class Nokia:
    def old_mobile(self):
        print("old mobile are only for phone calls")
class Iphone(Nokia):
    def new_phone(self):
        print("new mobiles are good for everthing")
s1=Iphone()
s1.old_mobile()
```

```python
#child constructor
class Parent:
    def __init__(self):
        print("this is parent constructor")
class Child(Parent):
    def __init__(self):
        print("this is child constructor")
```

```python
#method overriding
class Parent:
    def work(self):
        print("i will go to work")
class Child(Parent):
    def work(self):
        print("i will go to school")
s1=Child()
s1.work()
```

```python
#super method
class Parent:
    def __init__(self):
        print("i am parent")
class Child(Parent):
    def __init__(self):
        super().__init__()
    print("i am child")
s1=Child()
```

```python
#ques
class Account:
    def __init__(self,balance):
        self.balance=balance
    def show_balance(self):
        print(self.balance)
class Saving(Account):
    def add_money(self,ammount):
        self.balance+=ammount
        print("Money added:",ammount)
        print("updated balance:",self.balance)
obj=Saving(50000)
obj.show_balance()

obj.add_money(500)
obj.show_balance()
```

# Diamond Problem 

```python
class A:
    print("A")
class B(A):
    print("B")
class C(A):
    print("C")
class D(B,C):
    print("D")
print(D.mro())
```
